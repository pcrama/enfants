```python
%matplotlib inline
import imp
from matplotlib import pyplot as plt
import math
from premier_jet import *
```

```python
L, H = 1, math.sqrt(3) / 2 
triangle = decoupe_polygone_convexe(image_pixelisee('dessin_voeux_couleurs.png', [0, 0], [L, H]), [[0, 0], [L, 0], [L / 2, H]])
montre(triangle, [0, 0], [L, H], 256)
```

```python
six_triangles = superpose(
    triangle,
    rotation(triangle, 1 * 60, cx=0, cy=0),
    rotation(triangle, 2 * 60, cx=0, cy=0),
    translation(rotation(triangle, 3 * 60, cx=0, cy=0), [-L/2, H]),
    translation(rotation(triangle, 4 * 60, cx=0, cy=0), [-3 * L/2, H]),
    translation(rotation(triangle, 5 * 60, cx=0, cy=0), [L/2, H]),
    )
montre(six_triangles, [-2 * L, 0], [1.5 * L, H], 512)
```

```python
plan = pavage_parallelogramme(six_triangles, [-2 * L, 0], [-3 * L / 2, H], [L, 0])
montre(plan, [-2.5 * L, -H/2], [2 * L, 3*H/2], 512)
```

```python
R = 0.275
sauve('voeux_dessines_dans_cercle.png',
      decoupe_circulaire(comprime_dans_un_cercle(plan, 0.5),
                         [0, 0],
                         R),
      [-R, -R],
      [R, R],
      2048), 
```
